```python
# Create a LSTM model to predict the next bit of a LFSR
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers


def create_data(num_samples=10000, initial_key1=None, initial_key2=None, ln=5):
    X, y = [], []
    for i in range(580):
        key = np.binary_repr(i, width=10)
        key = [int(k) for k in key]
        key1 = key[:5]
        key2 = key[5:]
        x = []
        for j in range(ln + 1):
            nxt1 = key1[0] ^ key1[1]
            nxt2 = key2[0] ^ key2[2]
            x.append(nxt1 ^ nxt2)
            key1 = np.roll(key1, -1)
            key2 = np.roll(key2, -1)
            key1[-1] = nxt1
            key2[-1] = nxt2
        X.append(x[:-1])
        y.append(x[-1])
    return np.array(X), np.array(y)


def create_val_data(ln=5):
    X, y = [], []
    for i in range(580, 1024):
        key = np.binary_repr(i, width=10)
        key = [int(k) for k in key]
        key1 = key[:5]
        key2 = key[5:]
        x = []
        for j in range(ln + 1):
            nxt1 = key1[0] ^ key1[1]
            nxt2 = key2[0] ^ key2[2]
            x.append(nxt1 ^ nxt2)
            key1 = np.roll(key1, -1)
            key2 = np.roll(key2, -1)
            key1[-1] = nxt1
            key2[-1] = nxt2
        X.append(x[:-1])
        y.append(x[-1])
    return np.array(X), np.array(y)
```

```python
ln = 10
X_train, Y_train = create_data(
    ln=ln,
)
X_val, Y_val = create_val_data(ln=ln)
```

```python
model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Dense(1000, input_shape=(ln,), activation="relu"),
        # tf.keras.layers.Embedding(2, 5, input_length=5),
        # tf.keras.layers.LSTM(5),
        tf.keras.layers.Dense(1, activation="sigmoid"),
    ]
)
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
model.fit(X_train, Y_train, epochs=100, validation_data=(X_val, Y_val))
```
