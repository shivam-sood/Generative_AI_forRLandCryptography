```python
# Create a LSTM model to predict the next bit of a LFSR
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers


def create_data(
    num_samples=10000, initial_key1=None, initial_key2=None, initial_key3=None, ln=5
):
    if initial_key1 is None:
        initial_key1 = np.random.randint(0, 2, 3)
        print("Initial key 1: ", initial_key1)
    if initial_key2 is None:
        initial_key2 = np.random.randint(0, 2, 7)
        print("Initial key 2: ", initial_key2)
    if initial_key3 is None:
        initial_key3 = np.random.randint(0, 2, 7)
        print("Initial key 3: ", initial_key3)

    # data = initial_key
    key1 = initial_key1
    key2 = initial_key2
    key3 = initial_key3
    x = []
    y = []
    data = []
    for _ in range(num_samples):
        nxt1 = key1[0] ^ key1[2]
        nxt2 = key2[0] ^ key2[6]
        nxt3 = key3[0] ^ key3[4]
        data.append((nxt1 & nxt2) ^ ((1 - nxt1) & nxt3))
        if len(data) >= ln + 1:
            x.append(data[-(ln + 1) : -1])
            y.append(data[-1])
        key1 = np.roll(key1, -1)
        key2 = np.roll(key2, -1)
        key3 = np.roll(key3, -1)
        key1[-1] = nxt1
        key2[-1] = nxt2
        key3[-1] = nxt3

    return np.array(x), np.array(y)
```

```python
for i in range(2**17):
    # key = np.binary_repr(i, width=17)
    # key1 = np.array([key[j] for j in range(3)]).astype(int)
    # key2 = np.array([key[j] for j in range(3, 10)]).astype(int)
    # key3 = np.array([key[j] for j in range(10, 17)]).astype(int)
    ln = 30
    X, Y = create_data(
        5000,
        ln=ln,
        # initial_key1=key1,
        # initial_key2=key2,
        # initial_key3=key3,
        # initial_key1=np.array([0, 0, 1]),
        # initial_key2=np.array([1, 1, 1, 0, 0, 0, 1]),
        # initial_key3=np.array([0, 0, 0, 1, 0, 0, 0]),
    )
    X_train, Y_train = X[: 1000 - ln], Y[: 1000 - ln]
    X_test, Y_test = X[1000 - ln :], Y[1000 - ln :]
    # print(X.shape)
    model = tf.keras.models.Sequential(
        [
            # tf.keras.layers.Dense(10000, input_shape=(ln,), activation="relu"),
            tf.keras.layers.Embedding(2, 128, input_length=30),
            # tf.keras.layers.LSTM(5),
            tf.keras.layers.Flatten(),
            tf.keras.layers.Dense(100, activation="relu"),
            tf.keras.layers.Dense(1, activation="sigmoid"),
        ]
    )
    model.compile(
        optimizer="adam", loss="binary_crossentropy", metrics=["binary_accuracy"]
    )
    # model.summary()
    model.fit(
        X_train,
        Y_train,
        epochs=300,
        # validation_data=(X_test, Y_test),
        batch_size=512,
        verbose=0,
    )
    model.evaluate(X_test, Y_test)
```
